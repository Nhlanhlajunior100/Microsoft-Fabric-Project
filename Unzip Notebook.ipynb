### Let's import the necessary modules to extract the data from the zip files

```python
# Importing the ZipFile class form zipfile module.
from zipfile import ZipFile

# The os module contains methods for interacting with the operating system. 
import os
```

```python
# Working directory
os.getcwd()
```

```python
#Getting a list of the files in the File API path
api_path = "/lakehouse/default/Files/NYC Green Taxi Data"
os.chdir(api_path) 
os.getcwd()
```

```python
zipped = os.listdir() 
zipped
```

```python
#creating a ZipFile object to extract the contents of the zip file to folder named extracted
zip_file = ZipFile(zipped[1])
```

```python
target_dir = "/lakehouse/default/Files/NYC Green Taxi Data/Bronze"
zip_file.extractall(target_dir) 
zip_file.close()
```
