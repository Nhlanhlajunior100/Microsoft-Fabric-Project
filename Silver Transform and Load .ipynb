```python
from pyspark.sql.types import *
import os
```

```python
#Creating a parameter cell to run the notebook from a pipeline
month= "month=01"
table_name = "Table_Defaut"
```

```python
Year = "2020"
```

```python
nyc_schema = StructType([
    StructField("VendorID", IntegerType()),
    StructField("lpep_pickup_datetime", TimestampType()),
    StructField("lpep_dropoff_datetime", TimestampType()),
    StructField("Store_and_fwd_flag", StringType()),
    StructField("RateCodeID", IntegerType()),
    StructField("PULocationID", IntegerType()),
    StructField("DOLocationID", IntegerType()),
    StructField("Passenger_count", IntegerType()),
    StructField("Trip_distance", DoubleType()),
    StructField("Fare_amount", DecimalType()),
    StructField("Extra", DecimalType()),
    StructField("MTA_tax", DecimalType()),
    StructField("Tip_amount", DecimalType()),
    StructField("Tolls_amount", DecimalType()),
    StructField("Ehail_fee",DecimalType()),
    StructField("Improvement_surcharge", DecimalType()),
    StructField("Total_amount", DecimalType()),
    StructField("Payment_type", IntegerType()),
    StructField("Trip_type", IntegerType()),
    StructField("Congestion_surcharge", DecimalType())
    #StructField("Year", IntegerType()),
    #StructField("Month", IntegerType())
    ])
```

```python
#importing a single csv from the bronze layer shorcut
#df = spark.read.format("csv").option("header", "true").option(schema = nyc_schema).load("Files/Bronze/raw/trip_data_green_csv/year=2020/"+ file_name)

#Using a parameter to import csv files from the bronze layer
df = spark.read.csv("Files/Bronze/raw/trip_data_green_csv/year=2020/"+ month, header = True, schema = nyc_schema)
#display(df)
```

```python
#Splitting date and time columns and renaming them appropriately
from pyspark.sql.functions import *

df = df.withColumn("Pickup_Date", to_date("lpep_pickup_datetime")).\
    withColumn("Pickup_Time", date_format("lpep_pickup_datetime", "HH:mm:ss")).\
    withColumn("Dropoff_Date", to_date("lpep_dropoff_datetime")).\
    withColumn("Dropoff_Time", date_format("lpep_dropoff_datetime", "HH:mm:ss"))
display(df)
```

```python
#Load the data into a delta table

df.write.format("delta").mode("append").saveAsTable(table_name)
```
